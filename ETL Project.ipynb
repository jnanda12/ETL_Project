```python
player_df = df[['full_name', 'age', 'position', 'minutes_played_overall', 'nationality', 'appearances_overall',\
   'goals_overall', 'assists_overall', 'yellow_cards_overall', 'red_cards_overall', 'rank_in_league_top_attackers',\
   'rank_in_league_top_midfielders', 'rank_in_league_top_defenders', 'rank_in_club_top_scorer']]

team_df = df1[['country', 'matches_played', 'wins', 'draws', 'losses', 'points_per_game', 'performance_rank',
    'goals_scored', 'goals_conceded', 'goal_difference']]
```

```python
player_df.columns = ['full_name', 'age', 'position', 'minutes_played', 'nationality',
                 'appearances', 'goals', 'assists', 'yellow_cards', 'red_cards',
                 'rank_in_league_top_attackers', 'rank_in_league_top_midfielders',
                 'rank_in_league_top_defenders', 'rank_in_club_top_scorer']
player_df.head()
```

```python
team_df = team_df.replace({'Ivory Coast':"Côte d'Ivoire"})
team_df.head(17)
```

```python
engine = create_engine(f"mysql://root:{password}@localhost:3306/soccer")
conn = engine.connect()
```

```python
engine.table_names()
```

```python
team_df.to_sql(name = 'team', con=engine, if_exists = 'append', index = False)
```

```python
pd.read_sql_query('select * FROM team', con=engine).head()
```

```python
player_df.to_sql(name = 'players', con = engine, if_exists = 'append', index = False)
```

```python
pd.read_sql_query('select * FROM players', con=engine).head()
```

```python
pd.read_sql_query('SELECT * FROM Players p JOIN Team t ON t.country = p.nationality', con=engine).head()
```
